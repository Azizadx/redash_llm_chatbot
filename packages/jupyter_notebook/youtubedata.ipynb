```python

# Read JSON data from the file
try:
    with open(json_data, 'r') as file:
        data = json.load(file)
    print("Data loaded successfully, type:", type(data))
except json.JSONDecodeError as e:
    print("Error decoding JSON: ", e)
    data = {}  # Set to an empty dictionary in case of error
except FileNotFoundError as e:
    print("File not found: ", e)
    data = {}  # Set to an empty dictionary in case of file not found

# Check if data is a dictionary
if not isinstance(data, dict):
    print("Error: Data is not a dictionary")
else:
    # Flatten the data and convert to DataFrame
    rows = []
    for channel_id, channel_info in data.items():
        channel_stats = channel_info.get('chnnel_statistics', {})
        video_data = channel_info.get('video_data', {})
        for video_id, video_info in video_data.items():
            video_info['videoId'] = video_id
            video_info['channelId'] = channel_id
            row = {**channel_stats, **video_info}
            rows.append(row)

    df = pd.DataFrame(rows)
df.head() # Display the first few rows of the DataFrame
```
